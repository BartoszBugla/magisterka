EDA jest niezbędnym narzędziem w pracy analityków danych, statystyków i naukowców zajmujących się danymi. - https://astrafox.pl/slownik/exploratory-data-analysis-eda/

## Generowanie mapy dla zadanego zbioru danych

```python
from folium.plugins import MarkerCluster
import folium

df = df.sample(3000)
map_center = [df['latitude'].mean(), df['longitude'].mean()]

m = folium.Map(location=map_center, zoom_start=7)

marker_cluster = MarkerCluster().add_to(m)

for index, row in df.iterrows():
    popup_text = f"<b>{row['name']}</b><br><i>{row['category']}</i>"
    
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=popup_text,
        tooltip=row['name']
    ).add_to(marker_cluster) 

m.save("interactive_map.html")
```
